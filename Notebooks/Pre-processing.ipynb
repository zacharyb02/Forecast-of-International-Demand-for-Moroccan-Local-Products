```python
import pandas as pd
```

```python
def merge_df(file_path_1,file_path_2,file_path_3,file_path_4,file_path_5):
    df_v_1 = pd.read_csv(file_path_1, delimiter='\t')
    df_v_2 = pd.read_csv(file_path_2, delimiter='\t')
    df_v_3 = pd.read_csv(file_path_3, delimiter='\t')
    df_v_4 = pd.read_csv(file_path_4, delimiter='\t')
    df_v_5 = pd.read_csv(file_path_5, delimiter='\t')
    merged_df = df_v_5.merge(df_v_4, on='Product label', how='outer') 
    merged_df = merged_df.merge(df_v_3, on='Product label', how='outer') 
    merged_df = merged_df.merge(df_v_2, on='Product label', how='outer') 
    merged_df = merged_df.merge(df_v_1, on='Product label', how='outer')
    merged_df = merged_df.iloc[:, :-1]
    return merged_df
```
